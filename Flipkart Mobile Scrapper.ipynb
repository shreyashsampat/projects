Data Extraction from Flipkart pages

```python
for i in range(1,16):
    Product_Name=[]
    Price=[]
    Product_Rating=[]
    Camera_Rating=[]
    Battery_Rating=[]
    Display_Rating=[]
    Design_Rating=[]
    specs=[]
    info={}
    req=['Model Name', 'Color','SIM Type', 'Touchscreen', 'Display Size', 'Resolution','Operating System','Processor Core', 'Primary Clock Speed', 'Internal Storage', 'RAM','Primary Camera Available','Primary Camera', 'Secondary Camera Available','Secondary Camera','Full HD Recording','Network Type','Bluetooth Support','Wi-Fi','Audio Jack','GPS Support','Battery Capacity','Weight','Domestic Warranty']
    mn=[]
    color=[]
    st=[]
    ts=[] 
    ds=[]
    res=[]
    os=[]
    pc=[]
    pcs=[]
    se=[] 
    ram=[]
    pca=[]
    pcam=[]
    sca=[]
    scam=[]
    fhd=[]
    nt=[]
    bs=[]
    wifi=[]
    aj=[]
    gps=[]
    bc=[]
    w=[]
    dw=[]
    my_list = [mn, color, st, ts, ds, res, os, pc, pcs, se, ram, pca, pcam, sca, scam, fhd, nt, bs, wifi, aj, gps, bc, w, dw]
    URL="https://www.flipkart.com/search?q=mobile+phones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&as-pos=1&as-type=HISTORY&page="+str(i)
    headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246"}
    r = requests.get(url=URL, headers=headers)
    soup=BeautifulSoup(r.content,'html5lib')
    divs=soup.findAll('div',attrs={'class':"_2kHMtA"})
    for dv in divs:
        pURL=dv.find('a',attrs={'class':'_1fQZEK'})
        product='https://flipkart.com'+pURL.attrs['href']
        r = requests.get(url=product, headers=headers)
        soup=BeautifulSoup(r.content,'html5lib')
        try:
            name=soup.find('span',attrs={'class':'B_NuCI'}).text
        except:
            name="NA"
        try:
            price=soup.find('div',attrs={'class':'_30jeq3 _16Jk6d'}).text
        except:
            price="0"
        try:
            overall_rating=soup.find('div',attrs={'class':'_3LWZlK'}).text
        except:
            overall_rating="0.0"
        #Extracting individual rating
        ratings=soup.findAll('text',attrs={'class':'_2Ix0io'})
        try:
            camera_rating=ratings[0].text
        except:
            camera_rating = 0.0
        try:
            battery_rating=ratings[1].text
        except:
            battery_rating = 0.0
        try:
            display_rating=ratings[2].text
        except:
            display_rating = 0.0
        try:
            design_rating=ratings[3].text
        except:
            design_rating = 0.0
        #Flushing into variables
        Product_Name.append(name)
        Price.append(price)
        Product_Rating.append(overall_rating)
        Camera_Rating.append(camera_rating)
        Battery_Rating.append(battery_rating)
        Display_Rating.append(display_rating)
        Design_Rating.append(design_rating)
        table=soup.findAll('table',attrs={'class':'_14cfVK'})
        for c in table:
            rows=c.prettify()
            s=BeautifulSoup(rows,'html5lib')
            p=s.text.split("\n")
            n = [c.strip() for c in p if c.strip()]
            for z in n:
                specs.append(z)
        key=[]
        v=[]
        for j in range(len(specs)):
            if (j%2==0):
                if specs[j].strip() != "":
                    key.append(specs[j])
                else:
                    key.append("NA")
            else:
                if specs[j].strip() != "":
                    v.append(specs[j])
                else:
                    v.append("NA")
        for x in range (len(v)):
            info.update({key[x]:v[x]})
        o=0
        for q in req:
            for j in range(o,len(req)):
                try:
                    my_list[j].append(info[q])
                except:
                    my_list[j].append('NA')
                o=o+1
                break
        print(i)
    print(len(Product_Name))
    Final={'Name':Product_Name, 'Price':Price, 'Model Name':mn,'Color':color,'SIM Type':st,'Touchscreen':ts,'Display Size':ds,'Resolution':res,'Operating System':os,'Processor Core':pc, 'Primary Clock Speed':pcs, 'Internal Storage':se, 'RAM':ram,'Primary Camera Available':pca,'Primary Camera':pcam, 'Secondary Camera Available':sca,'Secondary Camera':scam,'Full HD Recording':fhd,'Network Type':nt,'Bluetooth Support':bs,'Wi-Fi':wifi,'Audio Jack':aj,'GPS Support':gps,'Battery Capacity':bc,'Weight':w,'Domestic Warranty':dw, 'Product Rating':Product_Rating, 'Camera Rating':Camera_Rating, 'Battery Rating':Battery_Rating, 'Display Rating':Display_Rating, 'Design Rating':Design_Rating}
    df=pd.DataFrame(Final)
    df.to_csv("Data/Page_"+i.__str__()+".csv",index=False)
```
